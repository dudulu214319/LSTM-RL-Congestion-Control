```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

mu = 200
sigma = 25
n_bins = 50
x = np.random.normal(mu, sigma, size=100)

fig, ax = plt.subplots(figsize=(8, 4))

# plot the cumulative histogram
n, bins, patches = ax.hist(x, n_bins, density=True, histtype='step',
                           cumulative=True, label='Empirical')

# Add a line showing the expected distribution.
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
y = y.cumsum()
y /= y[-1]

ax.plot(bins, y, 'k--', linewidth=1.5, label='Theoretical')

# # Overlay a reversed cumulative histogram.
# ax.hist(x, bins=bins, density=True, histtype='step', cumulative=-1,
#         label='Reversed emp.')

# tidy up the figure
ax.grid(True)
ax.legend(loc='right')
ax.set_title('Cumulative step histograms')
ax.set_xlabel('Annual rainfall (mm)')
ax.set_ylabel('Likelihood of occurrence')

plt.show()
```

```python
import random
import matplotlib.pyplot as plt
from matplotlib.collections import EventCollection
import numpy as np
import matplotlib.ticker as ticker
import mpl_toolkits.axisartist as axisartist

'''生成fake_receiving_rate.bottleneck = 100, loss = 200'''
def fake_RR(bottleneck=50, loss=100):
    x = 0
    flag = 0
    while flag<bottleneck:
        rr = 0.85 * x + random.uniform(-2,2)
        x += 1
        flag +=1
        yield rr
    rr_bott = rr
    # rr = bottlenec * 1.3
    while flag<loss:
        rr = rr_bott + random.uniform(-2,2)
        flag +=1
        yield rr

'''生成fake_RTT.bottleneck = 100, loss = 200'''
def fake_RTT(bottleneck=50, loss=100):
    x = 0
    flag = 0
    min_RTT = 40
    while flag<bottleneck:
        rtt = min_RTT + random.uniform(-2,2)
        flag +=1
        yield rtt
    while flag<loss:
        rtt = x + random.uniform(-2,2) + min_RTT
        x += 1
        flag +=1
        yield rtt

if __name__ == '__main__':
    bottleneck = 25
    loss = 50
    count = 0
    Rr_list = []
    RTT_list = []
    
    Rr = fake_RR(bottleneck, loss)
    RTT = fake_RTT(bottleneck, loss)
#     next(Rr)
    while True:
        try:
            Rr_list.append(next(Rr))
            RTT_list.append(next(RTT))
            count += 1
        except:
            break
    Rr_list[0] = 0
    RTT_list[0] = 40

    # Fixing random state for reproducibility
    np.random.seed(19680801)
    xdata = np.arange(loss)

    print(len(Rr_list))
    print(len(RTT_list))
    assert(len(Rr_list) == len(xdata))
    assert(len(RTT_list) == len(xdata))
    
    # transform the data
    xdata = np.array(xdata) / 5
    Rr_list = np.array(Rr_list) / 5
    RTT_list = np.array(RTT_list) * 2.5

    # plot the data
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    
#     ax1.axis[:].set_visible(True)
    ax1.grid(ls = "--", lw = 0.5, color = "#4E616C", axis='y')
    lns1 = ax1.plot(xdata, Rr_list, 'k', linewidth=1.5, label='Receiving Rate', color='r', marker = "o", mfc = "white", ms = 5)
    lns2 = ax2.plot(xdata, RTT_list, 'k', linewidth=1.5, label='Delay', color='b', marker = "o", mfc = "white", ms = 5)
    
#     fig.add_axes(ax1)
#     fig.add_axes(ax2)
    
#     ax.xaxis.set_major_locator(ticker.MultipleLocator(2)) # ticker every 2 matchdays
#     xticks_ = ax.xaxis.set_ticklabels([x - 1 for x in range(0, loss + 3, 2)])
    
#     ax.xaxis.set_tick_params(length = 2, color = "#4E616C", labelcolor = "#4E616C", labelsize = 10)
#     ax.yaxis.set_tick_params(length = 2, color = "#4E616C", labelcolor = "#4E616C", labelsize = 10)    
    ax1.set_ylabel('Receiving Rate (mbps)')
    ax1.set_xlabel('Sending Rate (mbps)')
    ax2.set_ylabel('Delay (ms)')
    ax2.set_ylim(90,165)
        
    lns = lns1+lns2
    labs = [l.get_label() for l in lns]
    ax1.legend(lns, labs, loc=4)
    ax1.vlines(4, 0, 5, linestyles ="dashed", colors ="k")
    ax1.text(1.4,4.5, 'bottleneck', fontsize=13, color='k')

    # display the plot
#     plt.show()
    plt.savefig("Bottleneck.pdf", dpi=300, format="pdf")
```
